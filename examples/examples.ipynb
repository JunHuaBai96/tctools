```python
import numpy as np
import matplotlib.pyplot as plt
from tctools import load_table, plot_table, plot_exp_datafile, table_to_excel
```

# Example 1: Loading tables and ploting property diagrams

```python
df = load_table('STAINLESS_STEEL_PHASE_FRACTIONS.DAT', sort='T', fill=0)  # load table, sort by the temperature column, fill NaN values with 0
df['T'] = df['T'] - 273.15  # Kelvin to Celsius

ax = plot_table(df, 'T')
ax.set_xlabel('Temperature (°C)')
ax.set_ylabel('Phase fraction')

plt.show()
```

```python
df = load_table('BEARING_STEEL_PHASE_COMPOSITIONS.DAT', sort='T')
df['T'] = df['T'] - 273.15

ax1 = plot_table(df, 'T', colpattern='W(FCC_A1#1,*[!FE])')  # Plot table by filtering columns following wildcard pattern (W(FCC_A1#1,"everything but FE"))
ax1.set_xlabel('Temperature (°C)')
ax1.set_ylabel('Phase composition (wt. fraction)')

ax2 = plot_table(df, 'T', colpattern='W(*,C)')
ax2.set_xlabel('Temperature (°C)')
ax2.set_ylabel('Phase composition (wt. fraction)')

plt.show()
```
